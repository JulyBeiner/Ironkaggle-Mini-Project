King County House Sale Price Analysis (2014-2015)
Overview
This project explores the house sale price data for King County, including Seattle, spanning from May 2014 to May 2015. Our mission is to analyze key factors that impact housing prices and build a predictive model to estimate house sale prices. This project was conducted over three days, simulating a real-world scenario of collaborative data analysis and machine learning.

Project Timeline
Day 1: Data Loading and Initial Exploration
Tasks:
Download the dataset and load it into Python.
Conduct preliminary data exploration to understand the structure and distribution of the data.
Create initial visualizations to identify patterns and relationships between variables.
Goal: Familiarize the team with the data and start hypothesizing which features might impact house prices.
Day 2: Data Analysis and Model Building
Tasks:
Clean and preprocess the dataset (handle missing values, scale features, etc.).
Conduct deeper exploratory data analysis (EDA) to determine the most impactful variables.
Develop and train a machine learning model (e.g., linear regression) for predicting house prices.
Goal: Create a baseline model and assess its performance using relevant metrics.
Day 3: Model Refinement and Presentation
Tasks:
Fine-tune the model for better accuracy and robustness.
Finalize visualizations to support model results.
Prepare a presentation summarizing the machine learning process and results.
Goal: Present findings and demonstrate the model's predictive capabilities.
Dataset Information
The dataset includes one year of house sale prices in King County, featuring 21 columns:

id: Unique identifier for each house.
date: Date of sale.
price: Sale price (target variable).
bedrooms, bathrooms: Number of bedrooms and bathrooms.
sqft_living, sqft_lot: Interior living space and land size.
floors: Number of floors in the house.
waterfront: Indicator for waterfront view.
view: Number of times the house was viewed.
condition, grade: Overall condition and grade of the house.
sqft_above, sqft_basement: Square footage above ground and in the basement.
yr_built, yr_renovated: Construction and renovation year.
zipcode: ZIP code of the property.
lat, long: Geographical coordinates.
sqft_living15, sqft_lot15: Living space and land size for the nearest 15 neighbors in 2015.
Machine Learning Process
Steps Taken:
Data Preprocessing:
Checked for and handled missing values.
Encoded categorical variables as needed.
Normalized and scaled numerical features.
Feature Selection:
Analyzed correlations and feature importance.
Model Building:
Built a regression model (e.g., Linear Regression, Random Forest) to predict house prices.
Evaluated model performance using metrics such as Mean Absolute Error (MAE) and R-squared.
Model Refinement:
Tuned hyperparameters for improved model accuracy.
Key Findings
Highlighted which features most significantly affect house prices (e.g., square footage, location, number of bedrooms).
Analyzed properties valued at $650K and above for targeted insights into higher-end real estate trends.
